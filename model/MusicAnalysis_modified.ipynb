```python
import os
import shutil
import pandas as pd

metadata_file_path = 'autotagging_moodtheme.csv'
source_folder_path = 'C:/Users/GAIS/mtg-jamendo-dataset/path_to_download'
destination_folder_path = 'C:/Users/GAIS/mtg-jamendo-dataset/classified_music_data'

metadata = pd.read_csv(metadata_file_path)

for index, row in metadata.iterrows():
    file_name = row['PATH']
    tag = row['TAGS']

    core_mood = tag.split('---')[-1] if '---' in tag else tag

    tag_folder_path = os.path.join(destination_folder_path, core_mood)

    if not os.path.exists(tag_folder_path):
        os.makedirs(tag_folder_path)

    source_file_path = os.path.join(source_folder_path, file_name)

    try:
        if os.path.exists(source_file_path):
            shutil.copy(source_file_path, tag_folder_path)
            print(f'File {file_name} copied successfully to {tag_folder_path}')
        else:
            print(f'File {file_name} does not exist in the source folder.')
    except Exception as e:
        print(f'Error copying file {file_name}: {str(e)}')
```
